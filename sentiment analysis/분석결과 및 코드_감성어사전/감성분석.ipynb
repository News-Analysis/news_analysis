```python
import pandas as pd
import csv
import jpype
from konlpy.tag import Okt
```

```python
a=pd.read_csv('C:/Users/wooji/Desktop/새 폴더 (3)/lexicon/polarity.csv')
table=dict()
with open('polarity.csv','r',-1,'utf-8') as polarity:
    next(polarity)
    
    for line in csv.reader(polarity):
        key =str()
        for word in line[0].split(';'):
            key +=word.split('/')[0]
        table[key]={"부정":line[3],'중립':line[4],'긍정':line[6]}
```

```python
neg=0
net=0
pos=0
result=[]
data=pd.read_csv("이재명_seoul.csv",encoding='utf-8-sig')
splitter=Okt()
for j in data["본문"]:
    k=splitter.morphs(j)
    for i in k:
        try:
            neg=neg+float(table[i]["부정"])
            net+=float(table[i]["중립"])
            pos+=float(table[i]["긍정"])
        except:
            continue
    print("긍정:",pos)
    print("부정:",neg)
    print("중립:",net)
    aa=pos/(pos+neg)
    if(aa>=0.58):
        print("결과:",aa)
        result.append("긍정")
    elif(0.58>aa>0.43):
        print("결과:",aa)
        result.append("중립")
    elif(0.43>=aa):
        print("결과:",aa)
        result.append("부정")
result=pd.DataFrame(data=result,columns=["분석결과"])
data["분석결과"]=result
data.to_csv("이재명_서울경제_result.csv",encoding='utf-8-sig')
```
