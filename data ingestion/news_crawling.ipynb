```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

## 서울경제 data

```python
seoul_data = pd.DataFrame({'언론사':[],
                       '기자':[],
                       '제목':[],
                       '내용':[]})
```

```python
bigkinds_data = pd.read_excel("NewsResult_20210202-20210502.xlsx")
```

```python
# 언론사, 기자, 제목, 내용
url_list = bigkinds_data['URL'][:2000]
reporter_list = bigkinds_data['기고자'][:2000]
```

```python
for url, reporter in zip(url_list, reporter_list):
    data = requests.get(url)
    reporter = reporter.split()[0]
    html = BeautifulSoup(data.text, "html.parser")
    
    title = html.select('title')[0].text
    news_text = html.select('div.article_view')[0].text.strip()
    
    seoul_data.loc[len(seoul_data)] = ['서울경제', reporter, title, news_text]
```

```python
seoul_data
```

```python
seoul_data.to_csv('C:\\Users\\RYU\\seoul_data.csv', header=False, index=False, encoding='utf-8-sig')
```

## 머니투데이 data

```python
mt_data = pd.DataFrame({'언론사':[],
                       '기자':[],
                       '제목':[],
                       '내용':[]})
```

```python
bigkinds_data2 = pd.read_excel("mt_20210202-20210502.xlsx")
```

```python
url_list2 = bigkinds_data2['URL'][413:1500]
reporter_list2 = bigkinds_data2['기고자'][413:1500]
```

```python
for url, reporter in zip(url_list2, reporter_list2):
        data = requests.get(url)
        reporter = reporter[:3]
        html = BeautifulSoup(data.text, "html.parser")
    
        title = html.select('title')[0].text
        news_text = html.select('div#textBody')[0].text.strip()

        mt_data.loc[len(mt_data)] = ['머니투데이', reporter, title, news_text]
        time.sleep(10)
```
