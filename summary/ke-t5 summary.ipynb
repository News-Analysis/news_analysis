```python
# 모델 깃허브 : https://github.com/AIRC-KETI/ke-t5
news = pd.read_csv('/content/drive/MyDrive/news/NewsData.csv')
loaded = tf.saved_model.load('/content/drive/MyDrive/news/t5.1.1.base.gin.ke.ke_v100_span_corruption.ke_t5_nikl_summary_mixture_equal')
infer = loaded.signatures["serving_default"]
```

```python
news.shape
```

```python
news = news.dropna()
```

```python
content = news['content']
summary = []

from tqdm import tqdm

for i in tqdm(range(len(content))):
    try:
      input_str_summary = content[i]
      x = tf.constant([input_str_summary])
      result = infer(x)
      summary.append([out.decode('utf-8') for out in result['outputs'].numpy()])
    except KeyError:
      pass

summary = pd.DataFrame(summary)
summary.to_csv('/content/drive/MyDrive/news/summary.csv')
```
